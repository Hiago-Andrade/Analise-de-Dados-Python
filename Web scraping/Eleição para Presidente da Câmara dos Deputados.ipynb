```python
soup = BeautifulSoup(html_votos, 'html.parser')   # parses = Análise do html
```

```python
# Lista do código html com a classe que há as informações de voto de cada deputado

soup.find_all("div", attrs={"class": "custom-representative"})  # find_all = encontra todas / "div" = encontra todas a divisões / attrs={"class": "custom-representative"} = que tem um atributo com a classe "custom-representative" 
```

```python
# Imprimindo informação do nome do deputado
# Data-Choice é o nome da variável que está armazenado o nome do voto declarado por cada candidato, procure essa variável lista do código anterior

for deputado in soup.find_all("div", attrs={"class": "custom-representative"}):
    print(deputado['data-choice'])
    
```

```python
# As informações com o nome do deputado, nome do partido e estado estão em uma tag de imagem (no código com nome = <img / e sobrenome = alt) 
# Colocando todos esses dados em uma lista

dados = []
for deputado in soup.find_all("div", attrs={"class": "custom-representative"}):
    dados_deputado = deputado.img['alt']     # Pega o nome e o cargo de cada deputado
    dados.append(dados_deputado)

print (dados)
```

```python
# Na lista do código anterior vemos que o nome, o partido e o estado estão todos juntos.
# Precisamos filtrar essas informações utilizando uma expressão regular.

dados = []
for deputado in soup.find_all("div", attrs={"class": "custom-representative"}):
    resultado_expressão_regular = re.search(r'.*Federal (.*)\((.*?)–(.*?)\)',deputado.img['alt'])  # (.*) = 1º grupo, pega a palavra entre a palavra "Federal" e o identificador "("     |     \((.*?)-(.*?)\) = 2º grupo e 3º grupo, pega os x's (xxx-xxx)
    dados.append(resultado_expressão_regular.group(1))   # 1º Grupo Nome do Deputado
    dados.append(resultado_expressão_regular.group(2))   # 2º Grupo Partido
    dados.append(resultado_expressão_regular.group(3))   # 3º Grupo Estado
    
print (dados)
```

```python
# Juntondo os passos anteriores em um só

dados = []
for deputado in soup.find_all("div", attrs={"class": "custom-representative"}):
    
    dados_deputado = []
    dados_deputado.append(deputado['data-choice']) #voto do deputado
    resultado_busca_nome = re.search(r'.*Federal (.*)\((.*?)–(.*?)\)',deputado.img['alt'])
    dados_deputado.append(resultado_busca_nome.group(1)) #nome do deputado
    dados_deputado.append(resultado_busca_nome.group(2)) #partido
    dados_deputado.append(resultado_busca_nome.group(3)) #estado
    dados_deputado.append(deputado.img['src']) #link da imagem do deputado
    dados.append(dados_deputado) 
```
